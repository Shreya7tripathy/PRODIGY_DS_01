```python
# AGE GROUP DISTRIBUTION IN WORLD POPULATION
import matplotlib.pyplot as plt

# data for age distribution
age_groups = ["0-14", "15-24", "25-34", "35-44", "45-54", "55-64", "65+"]
age_percentages = [25, 16, 20, 12, 9, 8, 10]

# bar chart
plt.figure(figsize=(8, 6))
plt.bar(age_groups, age_percentages, color=['skyblue', 'lightgreen', 'orange', 'purple', 'gold', 'red', 'gray'])
plt.xlabel("Age Groups")
plt.ylabel("Percentage of Population")
plt.title("Age Distribution of a Population")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Display the bar chart
plt.show()
```

```python
# Create the histogram
plt.figure(figsize=(8, 6))
plt.bar(age_groups, age_percentages, color='skyblue', edgecolor='black')
plt.xlabel("Age Group")
plt.ylabel("Percentage of Population")
plt.title("Distribution of Age Groups in a Population")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
