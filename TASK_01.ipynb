|  AGE GROUPS  |  WORLD POPULATION PERCENTAGE  |
|--------------|-------------------------------|
|  0-14 Age    |           25 %                |
|  15-64 Age   |           65 %                |
|  65+ Age     |           10 %                |




```python
# BAR CHART FOR AGE GROUP DISTRIBUTION IN WORLD POPULATION
import matplotlib.pyplot as plt

# data for age distribution in world
age_groups = ["0-14", "15-64", "65+"]
age_percentages = [25, 65, 10]

# bar chart
plt.figure(figsize=(5, 6))
plt.bar(age_groups, age_percentages, color=['orange', 'purple', 'red'])
plt.xlabel("Age Groups")
plt.ylabel("Percentage of Population")
plt.title("Age Distribution of World Population")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Display the bar chart
plt.show()
```

```python
# HISTOGRAM FOR AGE GROUP DISTRIBUTION IN WORLD POPULATION 
import matplotlib.pyplot as plt

# Data for age groups and percentages
age_groups = ["0-14", "15-64", "65+"]
age_percentages = [25, 65, 10]

# histogram
plt.figure(figsize=(5, 6))
plt.bar(age_groups, age_percentages, color=["lightgreen", "gold", "coral"])
plt.xlabel("Age Group")
plt.ylabel("Percentage")
plt.title("Age Distribution")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```
